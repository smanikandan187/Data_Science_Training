```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(1,9)
y = np.random.randint(10,100,size=8)
data = np.random.randn(100)

print(x)
print(y)
```

```python
plt.figure(figsize=(6,4))
plt.title("Line Plot")
plt.plot(x,y,marker="*",color="Blue",linestyle="-.")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(5,4))
plt.scatter(x,y,color="Brown")
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
plt.figure(figsize=(5,4))
plt.bar(x,y,color="Green")
plt.title("Bar Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.barh(x,y,color="Orange")
plt.title("Horizontal Bar Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
